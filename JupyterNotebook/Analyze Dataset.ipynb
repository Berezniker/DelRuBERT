```python
with open("./Биология/Копия ych1.txt") as f:
    text = map(str.strip, f.readlines())
    text = filter(None, text)
    text = map(str.split, text)
    text = sum(text, [])
    text = list(text)
```

```python
len(list(text))
```

```python
len(set(text))
```

```python
with open("./Математика/1.txt") as f:
    text_m = map(str.strip, f.readlines())
    text_m = filter(None, text_m)
    text_m = map(str.split, text_m)
    text_m = sum(text_m, [])
    text_m = list(text_m)
```

```python
len(list(text_m))  # кол-во слов
```

```python
len(set(text_m))  # кол-во уникальных слов
```

```python
list(sorted(set(text_m), key=len, reverse=True))[:50]
```

```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("sberbank-ai/sbert_large_nlu_ru")
```

```python
tokenizer.tokenize('лексикографический')
# ['лекси', '##ко', '##графи', '##чески', '##и']
# лекс | ико | граф | ическ | ий
```

```python
tokenizer.tokenize('интегрировать')
# интегрировать
# интегр | ир | ова | ть
```

```python
tokenizer.tokenize('физиолого-биохимическому')
```

```python
tokenizer.tokenize('интегрированных')
# ['интегри', '##рованных']
#  интегр | ир | ова | нн | ый
```

```python
list(map(
    lambda word: (word, tokenizer.tokenize(word)),
    sorted(set(text_m), key=len, reverse=True),
))[:50]
```
